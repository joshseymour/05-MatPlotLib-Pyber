```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
merge_data = pd.merge(ride_data, city_data, on="city")

# Display the data table for preview
merge_data
```

```python
city_grouped = merge_data.groupby("city").agg({'fare':'mean',
                                               'ride_id':'count',
                                               'driver_count':'max',
                                               'type': 'max'
                                                })
city_grouped.head()
```

```python
city_renamed = city_grouped.rename(columns={
    "fare": "Avg_Fare", "ride_id": "Rider_Count", "driver_count": "Drivers", "type": "Type"
})
color_dict = {'Urban': "Coral", 'Suburban': "SkyBlue", 'Rural': "gold"}
city_renamed['Color'] = city_renamed['Type'].map(color_dict)
city_renamed
```

## Bubble Plot of Ride Sharing Data

```python
plt.scatter(city_renamed.Rider_Count, city_renamed.Avg_Fare, s=city_renamed.Drivers*10, c=city_renamed.Color, alpha= .7)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.text(42,36,"Note:")
plt.text(42,35,"The size of the dot correlates to the driver count, per city")
plt.grid(True)
plt.legend(city_renamed.Type)
plt.savefig("Pyber Ride Sharing Data (2016).png")
plt.show()
```

```python
city_pies = city_renamed.groupby("Type").agg({'Avg_Fare':'sum',
                                               'Rider_Count':'count',
                                               'Drivers':'sum',
                                               })
city_pies
```

```python
city_pies['Avg_Fare'] = city_pies['Avg_Fare'].astype(int)
city_pies.dtypes
```

```python
city_pies['p_fare'] = round(city_pies['Avg_Fare']/city_pies['Avg_Fare'].sum() * 100,2)
city_pies['p_rider'] = round(city_pies['Rider_Count']/city_pies['Rider_Count'].sum() * 100,2)
city_pies['p_driver'] = round(city_pies['Drivers']/city_pies['Drivers'].sum() * 100,2)
city_pies
```

```python
explode = (0.1,0,0)
colors = ["gold","SkyBlue","Coral"]
plt.pie(city_pies.p_fare, explode=explode, labels=city_pies.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Fares, by City (2016)")
plt.axis("equal")
plt.savefig("Percent_Fare.png")
plt.show()
```

```python
#explode = (0.1,0,0)
#colors = ["gold","SkyBlue","Coral"]
plt.pie(city_pies.p_rider, explode=explode, labels=city_pies.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Riders, by City (2016)")
plt.axis("equal")
plt.savefig("Percent_Riders.png")
plt.show()
```

```python
explode = (0,0,0.1)
#colors = ["gold","SkyBlue","Coral"]
plt.pie(city_pies.p_driver, explode=explode, labels=city_pies.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Percentage of Drivers, by City (2016)")
plt.axis("equal")
plt.savefig("Percent_Drivers.png")
plt.show()
```
